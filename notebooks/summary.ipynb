```python
# Function to compute summary statistics of a DataFrame
def compute_mean(df, name):
    """
    Calculate and print summary statistics for the given DataFrame.
    Parameters:
        df (pd.DataFrame): The DataFrame to analyze.
        name (str): Name of the dataset for context in output.
    """
    print(f"Mean Summary Statistics for {name}:")
    print("-" * 50)
    print("Mean:\n", df.mean(numeric_only=True))
    print("\n\n")
```

```python
# Compute mean for each dataset of sierra leone
compute_mean(sierralone_df, "Sierra Leone")
```

```python
# Compute mean for each dataset of benin
compute_mean(benin_df, "Benin")
```

```python
# Compute mean for each dataset of togo
compute_mean(togo_df, "Togo")
```

```python
# select higher mean value from the dataset between togo, benin and sierra leone
def select_higher_mean(df1, df2, df3, names):
    """
    Select the name of the dataset with the highest mean value.
    Parameters:
        df1 (pd.DataFrame): The first DataFrame to compare.
        df2 (pd.DataFrame): The second DataFrame to compare.
        df3 (pd.DataFrame): The third DataFrame to compare.
        names (list): List of names corresponding to the datasets [name1, name2, name3].
    Returns:
        str: The name of the dataset with the highest mean value.
    """
    means = [df1.mean(numeric_only=True).mean(), df2.mean(numeric_only=True).mean(), df3.mean(numeric_only=True).mean()]
    max_index = means.index(max(means))
    return names[max_index]
```

```python
# Select the dataset with the highest mean value
# necessary columns to be selected
relevant_columns = ['GHI', 'DNI', 'DHI', 'Tamb', 'RH', 'Precipitation', 'Cleaning']
# names of the datasets
names = ['Togo', 'Benin', 'Sierra Leone']
```

```python
# call the function to select the dataset with the highest mean value
highest_mean_df = select_higher_mean(togo_df[relevant_columns], benin_df[relevant_columns], sierralone_df[relevant_columns], names)

print(f"The dataset with the highest mean value is: {highest_mean_df}")
```

```python
# select higher median value from the dataset between togo, benin and sierra leone
def select_higher_median(df1, df2, df3, names):
    """
    Select the name of the dataset with the highest median value.
    Parameters:
        df1 (pd.DataFrame): The first DataFrame to compare.
        df2 (pd.DataFrame): The second DataFrame to compare.
        df3 (pd.DataFrame): The third DataFrame to compare.
        names (list): List of names corresponding to the datasets [name1, name2, name3].
    Returns:
        str: The name of the dataset with the highest median value.
    """
    medians = [df1.median(numeric_only=True).mean(), df2.median(numeric_only=True).mean(), df3.median(numeric_only=True).mean()]
    max_index = medians.index(max(medians))
    return names[max_index]
```

```python
# Select the dataset with the highest median value
highest_median_df = select_higher_median(togo_df[relevant_columns], benin_df[relevant_columns], sierralone_df[relevant_columns], names)
print(f"The dataset with the highest median value is: {highest_median_df}")
```

```python
# select higher standard deviation value from the dataset between togo, benin and sierra leone
def select_higher_std(df1, df2, df3, names):
    """
    Select the name of the dataset with the highest standard deviation value.
    Parameters:
        df1 (pd.DataFrame): The first DataFrame to compare.
        df2 (pd.DataFrame): The second DataFrame to compare.
        df3 (pd.DataFrame): The third DataFrame to compare.
        names (list): List of names corresponding to the datasets [name1, name2, name3].
    Returns:
        str: The name of the dataset with the highest standard deviation value.
    """
    stds = [df1.std(numeric_only=True).mean(), df2.std(numeric_only=True).mean(), df3.std(numeric_only=True).mean()]
    max_index = stds.index(max(stds))
    return names[max_index]
```

```python
# Select the dataset with the highest standard deviation value
highest_std_df = select_higher_std(togo_df[relevant_columns], benin_df[relevant_columns], sierralone_df[relevant_columns], names)
print(f"The dataset with the highest standard deviation value is: {highest_std_df}")
```
